#4. What is the average duration of movies, and how does it vary across different ratings?

```python
data1 = data[data['type'] == 'Movie']
data1['duration'] =  pd.to_numeric(data1['duration'].astype(str).str.extract('(\d+)')[0], errors='coerce').fillna(0).astype(int)
data1
```

```python
#average duration of movies
data1['duration'].mean()
```

```python
#average duration of movies across different ratings
avg= data1.groupby('rating')['duration'].mean()
avg
```

```python
plt.title('Average Duration of Movies by Rating')
avg.plot(kind='bar')
plt.show()
```
