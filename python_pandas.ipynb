```python
import pandas as pd
import seaborn as sns
```

```python
# rename the column name
data.columns
```

```python
data = data.rename(columns={'started':'started_year','Youtuber':'youtuber_name'})
```

```python
data.columns
```

```python
#identifying outliners using IQR
import numpy as np
data['started_year'].isna().sum()
```

```python
started = data['started_year'].dropna()
```

```python
q1 =started.quantile(0.25)
q2 =started.quantile(0.50)
q3 =started.quantile(0.75)

iqr = q3 - q3

lower_bound =q1 - 1.5 * iqr
upper_bound =q3 + 1.5 * iqr
```

```python
outliners = data[(data['started_year'] < lower_bound) | (data['started_year'] > upper_bound)]

print('outliners',outliners[['started_year']])
```

```python
# how to create new column in exsting dataset :

data['stage'] = data['started_year'].apply(lambda x: 'senior' if x>=2022 else 'junior')
print(data[['started_year','stage']].head())
```

```python
print(data[['started_year','stage']].tail(10))
```

```python
# group by and aggregation

data_groupby = data.groupby('category').agg({'started_year':['mean','max','min']})
print(data_groupby)
```

```python
# pivoting data
data_pivot = data.pivot_table(index='category',columns='stage',values='started_year',aggfunc='mean')
print(data_pivot)
```

```python
# memory optimization(downcasting)

data['started_year'] = data['started_year'].astype('int16')

data['rank'] = pd.to_numeric(data['rank'], downcast='integer')

print(data.dtypes)
```
