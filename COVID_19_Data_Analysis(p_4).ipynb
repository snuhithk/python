```python
# finding that dataset is contains any null values
data.isna().any()
```

```python
# finding the count of null values in each column
data.isna().sum()
```

```python
#filling the null values
data['Province/State'].fillna(data['Province/State'].mode()[0],inplace=True)
data['min'].fillna(data['min'].mean(),inplace=True)
data['max'].fillna(data['max'].mean(),inplace=True)
data['slp'].fillna(data['slp'].mean(),inplace = True)
data['dewp'].fillna(data['dewp'].mean(),inplace = True)
data['rh'].fillna(data['rh'].mean(),inplace = True)
data['ah'].fillna(data['ah'].mean(),inplace = True)
```

```python
data.isna().sum()
```

#2. Which countries or regions have the highest and lowest total confirmed cases and fatalities?

```python
data2=data.groupby('Country/Region')[['ConfirmedCases','Fatalities']].sum()
data2
```

```python
#for highest
dt=data2.sort_values('ConfirmedCases',ascending=False)
dt.head()
```
