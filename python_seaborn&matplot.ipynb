```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data =sns.load_dataset('tips')
```

```python
# bisic plot
data = sns.load_dataset('tips')
data.head()
```

```python
plt.title('Line Plot')
sns.lineplot(x='total_bill',y='tip',data=data)
plt.figure(figsize=(10,5))
plt.show()
```

```python
# bar plot

plt.title('Bar Plot')
sns.barplot(x='total_bill',y='tip',data=data)
plt.figure(figsize=(10,5))
plt.show()
```

```python
# scatter plot
plt.title('this is a Scatter Plot')
sns.scatterplot(x='total_bill',y='tip',hue ='sex',data=data)
plt.figure(figsize=(10,5))
plt.show()
```

```python
# univariant analysis
#histogram
plt.title('Histogram')
sns.histplot(data['total_bill'],bins=15,kde =True)
plt.figure(figsize=(10,5))
plt.show()
```

```python
plt.title('Box Plot')
sns.boxplot(y='total_bill',data=data)

plt.show()
```

```python

plt.title('Violin Plot')
sns.violinplot(y='total_bill',data=data)

plt.show()
```

```python
# count plot
plt.title('Count Plot')
sns.countplot(x='day',data=data)

plt.show()
```

```python
# kde : kernel densisty estimator

sns.kdeplot(data['total_bill'])
```

```python
# bi variant analysis:

#scatter plot

sns.scatterplot(x='total_bill',y='tip',data=data,hue ='day')
```

```python
#box plot
sns.boxplot(x='day',y='total_bill',data=data)
```

```python
#violin plot

plt.title('Violin Plot')
sns.violinplot(x='day',y='total_bill',data=data)

plt.show()
```

```python
# multi variant plot

#pair plot

sns.pairplot(data)
```

```python
sns.pairplot(data,hue = 'sex')
```

```python
# joint plot

sns.jointplot(x='total_bill',y='tip',data=data, kind='hex')
```

```python
data.drop(['sex','smoker','day','time'],axis=1,inplace=True)
data
```

```python
# correlation matrix and heatmap

plt.figure(figsize=(10,5))
corr = data.corr()
sns.heatmap(corr,annot=True,cmap='mako')
```

```python
data1=sns.load_dataset('titanic')
data1.head()
```

```python
# create customize plot in matplotlib

plt.figure(figsize=(10,5))
sns.scatterplot(x='age',y ='fare',hue='sex',data=data1)
plt.xlabel('age')
plt.ylabel('fare')
plt.title('Scatter Plot')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='age',y ='fare',hue='sex',data=data1)
plt.xlabel('age')
plt.ylabel('fare')
plt.legend(title='Gender')
plt.show()
```

```python
# time series visualization
# line plot
# best example is stock market
dataset = sns.load_dataset('flights')
dataset.head()
```

```python
sns.lineplot(x='year',y='passengers',data=dataset)
# light blue in graph is for densisty
```

```python
#time forcating - time series
# moving average - works on the error data

dataset['passengers'].rolling(window=12).mean().plot()
```

```python
dataset['rolling_data']=dataset['passengers'].rolling(window=12).mean()
```

```python
sns.lineplot(x='passengers',y='rolling_data',data=dataset,label ='passengers')
sns.lineplot(x='year',y='rolling_data',data=dataset,label ='year')

plt.legend()
```
