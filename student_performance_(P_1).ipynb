#2.Is there a significant relationship between class attendance percentage and exam scores?

```python
plt.title('lineplot')
sns.lineplot(x='Attendance',y='Exam_Score',hue='Gender',data=data)
plt.xlabel("Class Attendance")
plt.ylabel("Exam Score (out of 100)")
```

# 3. Does the average number of sleep hours per night influence exam performance?

```python
sns.jointplot(x='Sleep_Hours',y='Exam_Score',hue='Gender',data=data)
plt.xlabel("Average Sleep Hours per Night")
plt.ylabel("Exam Score (out of 100)")
```

# 4. How does family income and parental education level impact the student's exam scores?

```python
#mapping

education = {'High School':1,'Bachelor':2,'Master':3,'PhD':4}
data['Parental_Education_Level'] = data['Parental_Education_Level'].map(education)

income = {'Low':1,'Medium':2,'High':3}
data['Family_Income'] = data['Family_Income'].map(income)
```

```python
#visual representation:
sns.pairplot(data[['Family_Income','Parental_Education_Level','Exam_Score']])
```

#5. Do private school students perform better in exams than public school students?

```python
#Do private school students perform better in exams than public school students?

sns.jointplot(x='School_Type',y='Exam_Score',data=data)
plt.xlabel("School Type")
plt.ylabel("Exam Score (out of 100)")
```
